```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Analise exploratoria do primeiro dataset (producao-de-plastico-global)

* atribuição do dataframe na variavel df1
* Informações sobre df1
* tratamento dos dados, exclusão dos dados faltantes e duplicados
* grafico
* considerações finais 

```python
df1 = pd.read_csv('1- producao-de-plastico-global.csv')
df1.info()
```

```python
df1= df1.dropna()
df1 = df1.drop_duplicates()
df1.info()
```

```python
sns.scatterplot(df1, x='Ano', y='Produção Anual de Plástico', hue='Produção Anual de Plástico', palette='viridis')
```

Como podemos observar no grafico, a produção de plastico tem um crescimento exponencial, em 1950, a produção anual era de 2 milhões de toneladas, crescendo para quase 460 milhões de toneladas em 2019.

Também conseguimos constatar um crescimento recente na ultima decada (2010-2019), onde a produção aumentou de 313 milhões para 460 milhões de toneladas, indicando uma grande demanda por plástico.

O aumento exponencial na produção de plpastico destaca diversas preocupações ambientais, principalmente pela poluição por microplasticos, isso reforça a necessidade de soluções sustentaveis de reciclagem para preservar o meio ambiente.


ref: [https://mundoeducacao.uol.com.br/quimica/microplasticos.htm]


## Analise do segundo dataset (participacao-despejo-residuo-plastico.csv)


* atribuição do dataframe na variavel df2
* Informações sobre df2
* tratamento dos dados, exclusão dos dados faltantes e duplicados
* grafico de pares 
* considerações finais 

```python
df2 = pd.read_csv('2- participacao-despejo-residuo-plastico.csv')
df2.info()
```

```python
df2 = df2.dropna()
df2 = df2.drop_duplicates()
df2.info()
```

```python
plt.figure(figsize=(15, 8))  
sns.barplot(data=df2, x='Código', y='Participação na emissão global de plásticos para o oceano')
plt.title('Participação na emissão global de plásticos para o oceano por Entidade')
plt.xticks(rotation=90, ha='right', fontsize=6)  
plt.tight_layout()  
plt.show()
```

O Grafico acima mostra a relação dos paises com suas respectivas participações na emissão global de plasticos, por serem varios países a visualização fica um pouco comprometida, por esse motivo resolvi "quebrar o grafico" em alguns pontos especificos que poderão ser vistos abaixo.

```python
top10 = df2.nlargest(10, 'Participação na emissão global de plásticos para o oceano')
plt.xticks(rotation=45)
sns.barplot(data=top10, x='Entidade', y='Participação na emissão global de plásticos para o oceano')
```

No top 5 maiores emissores de plasticos, todos os países são da Asia. Não só isso, como cerca de 80% de toda a emissão de plastico vem da Ásia. O Brasil ocupa a sexta posição no ranking, sendo o unico país da ámerica do sul no top 10. 

## Analise do terceiro dataset (3 - destino-plastico.csv)


* atribuição do dataframe na variavel df3
* Informações sobre df3
* tratamento dos dados, exclusão dos dados faltantes e duplicados
* graficos
* considerações finais

```python
df3 = pd.read_csv('3- destino-plastico.csv')
df3.info()
```

```python
#Remoção da coluna ' Código' que não será utilizada
df3 = df3.drop(columns=[' Código'], axis=1)

```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df3, x="Entidade", y=" Participação da reciclagem do lixo total regional")
plt.title("Comparação da Participação da Reciclagem do Lixo Total Regional dentre todas as Regiões")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df3, x="Entidade", y=" Participação do lixo descartado e mal gerido do total regional")
plt.title("Comparação da Participação do Lixo Descartado e Mal Gerido entre Regiões")
plt.xticks(rotation=45)
plt.show()
```

Embora a China e a India sejam os países que mais reciclam, eles também conseguem ser os países que pior gerem o lixo, isso se dá provavelmente pelo grande acumulo de plastico na região. 

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df3, x="Entidade", y=" Participação da queima do lixo total regional")
plt.title("Comparação da Participação da Queima do Lixo Total Regional entre Regiões")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(14, 8))
sns.boxplot(data=df3, x="Entidade", y=" Participação do lixo encaminhado para aterros do total regional")
plt.title("Comparação da Participação do lixo encaminhado para aterros do total regional entre Regiões")
plt.xticks(rotation=45)
plt.show()
```
